```python
def calc_position(n, r):
    gap_theta = 2*np.pi / n
    theta = [gap_theta*n for n in range(n)]

    x = [r*np.cos(th) for th in theta]
    y = [r*np.sin(th) for th in theta]
    
    return x, y, theta
```

```python
calc_position(4, 3)
```

```python
def draw_pos(x, y):
    plt.scatter(x, y)
    plt.axis('equal')
    plt.show()
```

```python
X, Y, theta = calc_position(4, 3)
draw_pos(X, Y)
```

```python
X, Y, theta = calc_position(10, 3)
draw_pos(X, Y)
```
